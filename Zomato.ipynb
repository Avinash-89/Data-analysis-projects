CLEANING DATA USING PYTHON (REMOVING /5 FROM RATE COLUMN)

```python
def handlerate(value):
    value =str(value).split('/')
    value= value[0];
    return float(value)
    
dataframe['rate']=dataframe['rate'].apply(handlerate)
print(dataframe.head())
```

CONCLUSION---- MAJORITY OF PEOPLE ORDER FROM DINING RESTAURANT

```python
dataframe.head()
```

```python

grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()

# Create a DataFrame from the grouped data
result = pd.DataFrame({'votes': grouped_data})
result.index.name = 'Type of restaurant'

# Plot the data
plt.plot(result.index, result['votes'], c="purple", marker="o")

# Add labels and title
plt.xlabel("Type of restaurant", color="red", fontsize=20)
plt.ylabel("votes", color="red", fontsize=20)
```

CONCLUSION ---- DINNING RESTAURANT RECIVES  HIGHEST VOTES FROM CUSTOMER
